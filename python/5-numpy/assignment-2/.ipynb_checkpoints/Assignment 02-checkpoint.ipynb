#### 1: View and add the dataset

```python
#Import the necessary library
import numpy as np 
```

```python
#Manually add the Summer Olympics, London 2012 dataset as arrays

country = np.array(['GBR','CHN','RUS','US','KOR','JPN','GER'])
Gold = np.array([29,38,24,46,13,7,11])
Silver = np.array([17,28,25,28,8,14,11])
Bronze = np.array([19,22,32,29,7,17,14])
```

#### Find the country with maximum gold medals

```python
#Use the argmax() method to find the highest number of gold medals
max_Gold = Gold.argmax()
```

```python
#Print the name of the country
country[max_Gold]
```

#### Find the countries with more than 20 gold medals

```python
#Use Boolean indexing technique to find the required output
print(country[Gold >20])
```

#### Evaluate the dataset and print the name of each country with its gold medals and total number of medals

```python
#Use a for loop to create the required output
for i in range(len(country)):
    print(f'{country[i]}, gold medal {Gold[i]}, Total medals{Gold[i] + Silver[i] + Bronze[i]}')
```
